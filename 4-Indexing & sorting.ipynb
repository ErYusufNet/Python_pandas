## Sorting multiple columns!

```python
houses = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
```

```python
houses
```

```python
houses.sort_values("price")
```

```python
houses.sort_values("price", ascending=False)
```

```python
houses.sort_values("price", ascending=True)
```

```python
houses.sort_values("bedrooms", ascending=False)
```

```python
#How we can sort multiple columns ?
houses.sort_values(["bathrooms", "bedrooms"], ascending=False)
```

## Sorting text columns !

```python
titanic
```

```python
titanic.sort_values("name")
#as you see on the buttom there is not anything with Z , there is V how we can solve this problem?
```

```python
titanic.sort_values("name", key=lambda col: col.str.lower()) 
```
