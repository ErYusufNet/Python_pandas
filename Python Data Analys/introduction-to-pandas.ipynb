```python
#How we can change the model year of 9 - Nissan?



new_model_year = 2015
car_sales.at[9, "Model Year"] = new_model_year
car_sales
```

```python
# How we can count how much L fuel used the car according to odometer and using fuel per 100km ?
#Then declaring this as column to data 


car_sales["Total fuel used"] = car_sales["Odometer (KM)"] /100 * car_sales["Fuel per 100km"]
car_sales
```

```python
#How we can add number of wheels to all cars in one time?


car_sales["Number of wheels"] = 4
car_sales
```

```python
#How we can add price column?

price = [1500, 1350, 850]
Phone_store["Price"] = price
Phone_store
```

```python
#How we can add colour ?

colour = ("White", "Black", "Blue")
Phone_store["Colour"] = colour
Phone_store
```
